```python
from collections import defaultdict
from scipy.spatial import Voronoi, voronoi_plot_2d
import cdd

class PolynomialTessellation:
    def __init__(self, mu, ratio = 0.5, max_dist = 20):
        self.Ratio = ratio
        max_dist = 20
        boundaries = []
        for combo in itertools.product(*(((-1.,1.),) * len(mu[0]))):
            boundaries.append(np.array(combo) * max_dist)

        mu = np.concatenate((mu, boundaries))
        self.Vor = Voronoi(mu)

        point_to_point = defaultdict(set)
        self.Vertex_to_point = defaultdict(set)
        self.Point_to_vertex = defaultdict(set)

        for r, (rps, rvs) in enumerate(zip(self.Vor.ridge_points, self.Vor.ridge_vertices)):
            for i in rps:
                point_to_point[i] |= set(j for j in rps if j != i)
                for j in rvs:
                    if j > -1 and i not in self.Vertex_to_point[j]:
                        self.Vertex_to_point[j].add(i)
                    if j > -1 and j not in self.Point_to_vertex[i]:
                        self.Point_to_vertex[i].add(j)

        self.PointBoundaries = []
        for i, p in enumerate(self.Vor.points):
            A = []
            b = []
            adj = []
            for a in point_to_point[i]:
                n = self.Vor.points[a] - p
                r = np.linalg.norm(n) / 2
                n /= 2*r
                A.append(n)
                b.append(r)
                adj.append(a)
            A = np.array(A)
            b = np.array(b)
            self.PointBoundaries.append((i,A,b,adj))

    def eval(self, x):
        relevant_points = set()
        for p_i, A, b, adj in self.PointBoundaries:
            p = self.Vor.points[p_i]
            d = A @ (x - p)
            if np.all(d <= b*self.Ratio):
                return -(x-p)@(x-p), -2*(x-p), 1
            if np.all(d <= b):
                if p_i not in relevant_points:
                    relevant_points.add(p_i)
                for i in range(len(b)):
                    if d[i] >= 0:
                        if adj[i] not in relevant_points:
                            relevant_points.add(adj[i])
        vertex_combos = set()
        for i in range(2, len(x)+2):
            for point_combo in itertools.combinations(relevant_points, i):
                vs = set.intersection(*(self.Point_to_vertex[p_i] for p_i in point_combo))
                vs = frozenset(vs)
                if len(vs) > 0 and vs not in vertex_combos:
                    vertex_combos.add(vs)
        
        anchor_cache = {}
        
        for combo in vertex_combos:
            points = set.intersection(*(self.Vertex_to_point[v] for v in combo))
            r = []
            for v_i in combo:
                for p_i in points:
                    key = (v_i, p_i)
                    if key in anchor_cache:
                        r.append(anchor_cache[key])
                    else:
                        p = self.Vor.points[p_i]
                        a = p + self.Ratio * (self.Vor.vertices[v_i] - p)
                        value = (
                            a,
                            -(a-p)@(a-p),
                            -2*(a - p)
                        )
                        r.append(value)
                        anchor_cache[key] = value
                        
            mat = []
            xs = []
            fs = []
            dfs = []
            for a,f,df in r:
                mat.append(np.concatenate(([1], a), axis = 0))
                xs.append(a)
                fs.append(f)
                dfs.append(df)
            mat = cdd.Matrix(mat, number_type = 'fraction')
            mat.rep_type = cdd.RepType.GENERATOR
            poly = cdd.Polyhedron(mat)
            bA = np.array(poly.get_inequalities(), dtype = np.float64)
            b = bA[:,0]
            A = -bA[:,1:]
            if np.all(A @ x <= b):
                coeffs, err = fit_poly(xs, fs, dfs, 2)
                #print(err)
                return eval_poly(x, coeffs), grad_poly(x, coeffs), len(relevant_points)
        raise ValueError
        return 1, np.zeros_like(x), len(relevant_points)
            


def generate_regions(mu, ratio = 0.9):
    max_dist = 20
    boundaries = []
    for combo in itertools.product(*(((-1.,1.),) * len(mu[0]))):
        boundaries.append(np.array(combo) * max_dist)
    
    mu = np.concatenate((mu, boundaries))
    vor = Voronoi(mu)
    
    ridge_combo_to_vertex = defaultdict(list)
    vertex_to_ridge = defaultdict(list)
    
    vertex_to_point = defaultdict(set)
    point_to_vertex = defaultdict(set)
    
    for r, (rps, rvs) in enumerate(zip(vor.ridge_points, vor.ridge_vertices)):
        for i in rps:
            for j in rvs:
                if j > -1 and i not in vertex_to_point[j]:
                    vertex_to_point[j].add(i)
                if j > -1 and j not in point_to_vertex[i]:
                    point_to_vertex[i].add(j)
                    
    for r, rvs in enumerate(vor.ridge_vertices):
        for v in rvs:
            if v > -1:
                vertex_to_ridge[v].append(r)
    
    for v, ridges in vertex_to_ridge.items():
        for combo in itertools.combinations_with_replacement(ridges, 2):
            ridge_combo_to_vertex[combo].append(v)
    #print(dict(ridge_combo_to_vertex))
    #ridge_to_vertex = [frozenset(v for v in rvs if v != -1) for rvs in vor.ridge_vertices]
    #vertex_combinations = set()
    #for rtv1 in ridge_to_vertex:
    #    for rtv2 in ridge_to_vertex:
    #        a = rtv1 & rtv2
    #        if len(a) > 0 and a not in vertex_combinations:
    #            vertex_combinations.add(a)
    
    #fast_vertex_combinations = set(frozenset(vs) for vs in ridge_combo_to_vertex.values())
    #slow_vertex_combinations = set(frozenset(vs) for vs in vertex_combinations)
    #print(slow_vertex_combinations - fast_vertex_combinations)
    
    vertex_to_point = defaultdict(set)
    
    for r, (rps, rvs) in enumerate(zip(vor.ridge_points, vor.ridge_vertices)):
        for i in rps:
            for j in rvs:
                if j > -1 and i not in vertex_to_point[j]:
                    vertex_to_point[j].add(i)
    
    region_vertices = []
    anchor_cache = {}
    vertex_combo = set(map(frozenset, ridge_combo_to_vertex.values()))
    print(len(vertex_to_ridge))
    print(len(vertex_combo))
    print(len(ridge_combo_to_vertex))
    print(1/0)
    
    for combo in ridge_combo_to_vertex.values():
        points = set.intersection(*(vertex_to_point[v] for v in combo))
        r = []
        for v_i in combo:
            for p_i in points:
                key = (v_i, p_i)
                if key in anchor_cache:
                    r.append(anchor_cache[key])
                else:
                    p = vor.points[p_i]
                    x = p + ratio * (vor.vertices[v_i] - p)
                    value = (
                        x,
                        -np.linalg.norm(x - p)**2,
                        -2*(x - p)
                    )
                    r.append(value)
                    anchor_cache[key] = value
        #print(len(r))
        region_vertices.append(r)
    print(len(anchor_cache))
    regions = []
    for rvs in region_vertices:
        mat = []
        xs = []
        fs = []
        dfs = []
        for x,f,df in rvs:
            mat.append(np.concatenate(([1], x), axis = 0))
            xs.append(x)
            fs.append(f)
            dfs.append(df)
        mat = cdd.Matrix(mat, number_type = 'fraction')
        mat.rep_type = cdd.RepType.GENERATOR
        poly = cdd.Polyhedron(mat)
        bA = np.array(poly.get_inequalities(), dtype = np.float64)
        b = bA[:,0]
        A = -bA[:,1:]
        scale = np.linalg.norm(A, axis = 1)
        b /= scale
        A = (A.T / scale).T
        
        coeffs, err = fit_poly(xs, fs, dfs, 2)
        #print(err)
        regions.append(((A, b), coeffs))
        #print(coeffs)
        
    return regions, vor

mu = np.array((
    [0,0.25],
    [0.5,0],
    [1,0.25]
))
#regions, vor = generate_regions(mu)
tess = PolynomialTessellation(mu)

plt.figure()
ax = plt.gca(projection = '3d')
ax.scatter(*tess.Vor.points.T, color = 'black')
ax.scatter(*tess.Vor.vertices.T, color = 'blue')



ORDERS = [1, 3, 6, 10, 15, 21, 28, 36, 45, 55, 66, 78, 91, 105, 120, 136, 153, 171, 190, 210, 231,
          253, 276, 300, 325, 351, 378, 406, 435, 465, 496, 528, 561, 595, 630, 666,
          703, 741, 780, 820, 861, 903, 946, 990, 1035, 1081, 1128, 1176, 1225, 1275]
```

```python
import math
refvec = [0, 1]

def clockwiseangle_and_distance(point):
    # Vector between point and the origin: v = p - o
    vector = [point[0]-origin[0], point[1]-origin[1]]
    # Length of vector: ||v||
    lenvector = math.hypot(vector[0], vector[1])
    # If length is zero there is no angle
    if lenvector == 0:
        return -math.pi, 0
    # Normalize vector: v/||v||
    normalized = [vector[0]/lenvector, vector[1]/lenvector]
    dotprod  = normalized[0]*refvec[0] + normalized[1]*refvec[1]     # x1*x2 + y1*y2
    diffprod = refvec[1]*normalized[0] - refvec[0]*normalized[1]     # x1*y2 - y1*x2
    angle = math.atan2(diffprod, dotprod)
    # Negative angles represent counter-clockwise angles so we need to subtract them 
    # from 2*pi (360 degrees)
    if angle < 0:
        return 2*math.pi+angle, lenvector
    # I return first the angle because that's the primary sorting criterium
    # but if two vectors have the same angle then the shorter distance should come first.
    return angle, lenvector
```
