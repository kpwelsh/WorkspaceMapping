```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
from scipy.optimize import minimize, LinearConstraint, NonlinearConstraint
```

```python
h = 0.0001
l = 0
r = 2
x = np.arange(0, 4, h)
mu = np.array((1.5, 2.5))
height = np.array((1./h, 1./h))

f0 = np.zeros_like(x)
for m, _h in zip(mu, height):
    f0[int((m)/h)] = _h

f00f = np.concatenate((f0, -np.array(list(reversed(f0)))), axis = 0)
xx = np.arange(0, 8, h)

for t in [1e-2, 2e-2, 1e-1]:
    f = np.zeros_like(f0)
    for i in range(0, 200):
        ai = np.sum(np.sin(xx * i * np.pi/4) * f00f) * h
        f += ai * np.exp(-t * (i * np.pi/4)**2) * np.sin(x * i * np.pi/4)

    ddf = f - 2
    df = np.cumsum(ddf) * h
    f = np.cumsum(df) * h

    plt.plot(x, ddf)


plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
```

```python
h = 0.01
x = np.arange(0, 4, h)

mu = np.array((1, 2, 3))
b = np.array((0,0,-0.5))
xx = np.column_stack([x for _ in mu])

s = 1
v = np.max(-((xx - mu)/s)**2 + b, axis = 1)

def cons_d2(x):
    return (x[2:] - 2 * x[1:-1] + x[:-2]) / (h **2)

def cost(x):
    return np.linalg.norm((x - v), ord = 2)

for max_ddf,c in zip([2, 1, 0], ['blue','orange','green']):
    max_ddf = max_ddf/s**2

    l_constraint = LinearConstraint(np.eye(len(v)), v, np.inf)
    constraint_d2f = NonlinearConstraint(cons_d2, -np.inf, max_ddf)
    res = minimize(cost, v.copy(), method = 'SLSQP', constraints=[l_constraint, constraint_d2f])
    f = res['x']
    plt.plot(x[1:-1], (f[2:] - 2 * f[1:-1] + f[:-2]) / (h **2), color = c)
    
    for m in [1.5, 2.75]:
        _x = m -1 / (max_ddf*s**2 + 2) 
        plt.plot([_x, _x], [-2, 10], color = c, linestyle = '--')
        _x = m + 1 / (max_ddf*s**2 + 2)
        plt.plot([_x, _x], [-2, 10], color = c, linestyle = '--')
    plt.plot([0, 4], [max_ddf*s**2, max_ddf*s**2], color = c, linestyle = '--')
    
plt.gca().axes.get_xaxis().set_visible(False)
plt.gca().axes.get_yaxis().set_visible(False)
```
