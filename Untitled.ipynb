```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook
from scipy.optimize import minimize, LinearConstraint
```

```python
h = 0.01
x = np.arange(-1,3,h)
mu = np.array((0., 1.))
XX = np.column_stack([x for _ in mu])

v = np.max(-(XX - mu)**2, axis = 1)
D2 = np.eye(len(v))
for i in range(1, len(D2) - 1):
    D2[i,i-1:i+2] = [1, -2, 1]
D2[0,:4] = [2, -5, 4, -1]
D2[-1,-4:] = [-1, 4, -5, 2]
D2 /= h**2

D1 = np.eye(len(v))
for i in range(1, len(D1) - 1):
    D1[i,i-1:i+2] = [-0.5, 0, 0.5]
D1[0,:3] = [-1.5, 2, -0.5]
D1[-1,-3:] = [-0.5, 2, -1.5]
D1 /= h

max_ddf = 2
res = minimize(lambda f: np.linalg.norm(f - v), 
                v.copy(), 
                method = 'SLSQP', 
                bounds = list(zip(v, np.full_like(v, np.inf))),
                constraints = [LinearConstraint(D2, -np.inf, max_ddf)]
              )
f = res['x']

#f = np.exp(f)

L = 1.9
traj = []

def update(x0):
    _f = 1 - f + L * (x - x0)**2
    plt.gca().clear()
    plt.scatter([x0], [0])
    i = np.argmin(_f)
    traj.append(x[i])
    plt.scatter([x[i]], [_f[i]])
    plt.plot(x, _f)
    plt.plot(x[:len(traj)], traj)
    plt.plot(x, x, color = 'green', linestyle = '--')
    plt.ylim([-1,3])

def init():
    global traj
    traj = []
    
ani = FuncAnimation(plt.gcf(), update, x, interval = 20, init_func = init)
plt.show()


```
